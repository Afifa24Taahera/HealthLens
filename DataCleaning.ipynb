```python
import pandas as pd
```

```python
#Load original dataset
raw_data= pd.read_csv('/content/heart_cleveland_upload.csv')

#Shows all the info of dataset
print('Information of raw')
raw_data.info()

#Shows statistical info.
raw_data.describe()
```

```python
#Displaying top 5 rows of raw_data
print('Original dataset')
raw_data.head(5)
```

```python
##Renaming technical column names to meaningful labels
raw_data.rename(columns={'cp': 'chest pain type', 'trestbps':'resting blood pressure','chol':'serum cholestoral(mg/dl)','fbs':'fasting blood sugar > 120 mg/dl',
                         'restecg':'resting electrocardiographic results','thalach':'maximum heart rate achieved','exang':'exercise induced angina','oldpeak':'Induced depression(exercise vs rest)',
                         'slope':'slope peak exercise','ca':'colored vessels','thal':'thalassemia','condition':'disease condition'}, inplace=True)
display(raw_data.head(2))
```

```python
#Check for missing values
raw_data.isnull().sum()
```

```python
#Check for duplicates in dataset.
print(raw_data.duplicated().sum())
```

*   Dataset has 14 columns and 297 entries.
*   There are no missing values.
*   There are no duplicate

```python
#Replacing/converting categorical variables to interpretable form.
raw_data.replace({'sex':{1:'male', 0:'female'},'chest pain type':{ 0: 'typical angina',1: 'atypical angina',2: 'non-anginal pain',3: 'asymptomatic'},
                  'fasting blood sugar > 120 mg/dl':{1:'true', 0:'false'},
                  'resting electrocardiographic results':{0:'normal',1:'wave abnormality',2:'left ventricular hypertrophy'},
                  'exercise induced angina':{1:'yes', 0:'no'},'slope peak exercise':{0:'upsloping',1:'flat',2:'downsloping'},'thalassemia':{0:'normal', 1:'fixed defect', 2:'reversable defect'}
                  ,'disease condition':{0:'no disease', 1:'disease'}},inplace=True)
display(raw_data.head(3))
```

```python
#Export the cleaned dataset
raw_data.to_csv('cleaned_data_heart.csv',index=False)
print('Exported successfully')
```
